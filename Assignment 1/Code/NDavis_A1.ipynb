Question 1:

Implement a class for n-sided polygons and a class for points in a Euclidean system, namely *polygon* and *point* respectively. For example, a 4-sided polygon can be defined by 4 points P1, P2, P3, P4, and
P1-P4 are each points of the form point(X,Y), and X and Y are coordinates on the X and Y axis, respectively. The edges are listed counterclockwise starting at the lower left: P1 to P2, P2 to P3, P3 to P4, and P4 to P1.  The polygon class should work for polygons of any number of edges and have a function perimeter that returns its perimeter (sum of the lengths of the edges).

```python
class point():
    def __init__(self, x1, y1):
        self.x = x1
        self.y = y1

    def distance(self, pointer):
        xDist = self.x - pointer.x
        yDist = self.y - pointer.y
        return math.sqrt(math.pow(xDist, 2) + math.pow(yDist, 2))
```

```python
class Polygon():
    def __init__(self, polyPoints):
        self.polyPoints = polyPoints

    def perimeter(self):
        perimeter = 0
        for i in range(1, len(self.polyPoints)):
            point1 = self.polyPoints[i-1]
            point2 = self.polyPoints[i]
            perimeter += point1.distance(point2)

            if i + 1 == len(self.polyPoints):
                perimeter += self.polyPoints[-1].distance(self.polyPoints[0])
        return perimeter
```

```python
numPoints = int(input("Enter number of points in polygon : "))

points = []

for i in range(0, numPoints):
    xValue = int(input("Input X value for point %d: " % (i+1)))
    yValue = int(input("Input Y value for point %d: " % (i+1)))

    tmp = point(xValue, yValue)
    points.append(tmp)

polygon = Polygon(points)
print(polygon.perimeter())
```
